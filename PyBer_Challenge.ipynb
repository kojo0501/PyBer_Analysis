### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_type_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
#total_rides = pd.Series([rides_type_count], index=[type], name='Total Rides')
#print(total_rides)
print(rides_type_count)
```

```python
# 2. Get the total drivers for each city type
driver_type_sum = city_data_df.groupby(["type"]).sum()["driver_count"]
#total_drivers = pd.Series([driver_type_sum], index=[type], name="Total Drivers")
print(driver_type_sum)
```

```python
#  3. Get the total amount of fares for each city type
fare_type_sum = pyber_data_df.groupby(["type"]).sum()["fare"]
print(fare_type_sum)
```

```python
#  4. Get the average fare per ride for each city type. 
fare_average = fare_type_sum / rides_type_count
print(fare_average)
type(fare_average)
```

```python
# 5. Get the average fare per driver for each city type. 
driver_fare_average = fare_type_sum / driver_type_sum
print(driver_fare_average)
#df = pd.DataFrame(driver_fare_average)
#df
```

```python
#  6. Create a PyBer summary DataFrame. 
#pd.concat([total_rides, total_drivers], axis=1)
#summary_dict = [{"Total Rides": rides_type_count,
#                "Total Drivers": driver_type_sum,
#                "Total Fares": fare_type_sum,
#                "Average Fare per Ride": fare_average,
#                "Average Fare per Driver": driver_fare_average
#                }]
#summary_df = pd.DataFrame(data=summary_dict)
#print(summary_df)
total_rides_df = pd.DataFrame(rides_type_count)
total_drivers_df = pd.DataFrame(driver_type_sum)
total_fares_df = pd.DataFrame(fare_type_sum)
af_ride_df = pd.DataFrame(fare_average)
af_driver_df = pd.DataFrame(driver_fare_average)

#total_rides_df.set_index("type")
#total_drivers_df.set_index("type")

#pyber_summary_df = pd.concat([total_rides_df, total_drivers_df, total_fares_df])
#pyber_summary_df

pyber_summary_df = total_rides_df.join(total_drivers_df)
pyber_summary_df = pyber_summary_df.join(total_fares_df, lsuffix="_left")
pyber_summary_df = pyber_summary_df.join(af_ride_df, lsuffix="_left")
pyber_summary_df = pyber_summary_df.join(af_driver_df, lsuffix="_left")
pyber_summary_df = pyber_summary_df.rename(columns={"ride_id":"Total Rides", "driver_count":"Total Drivers", "fare":"Total Fares","0_left":"Average Fare per Ride", "0":"Average Fare per Driver"})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# A dictionary of high schools
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
rides_date_df = pyber_data_df.groupby(["date", "type"]).sum()["fare"]
rides_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
rides_date_df = rides_date_df.reset_index()
rides_date_df
#rides_date_df.dtypes
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_rides_date = rides_date_df.pivot(index = "date", columns = "type", values = "fare")
pivot_rides_date
#type(pivot_rides_date)
#pivot_rides_date.dtypes
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
# pivot_rides_date.loc(pivot_rides_date["date"] < '2019-04-29 00:00:00')
#date_loc = pivot_rides_date.loc(pivot_rides_date["date"] < '2019-04-29', "date")
#pivot_rides_date["date"]
date_range_df = pivot_rides_date.loc['2019-01-01':'2019-04-28']
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
date_range_df.index = pd.to_datetime(date_range_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
date_range_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_totals_df = date_range_df.resample('W').sum()
weekly_totals_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
#import numpy as np
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
#dates = ["Jan", "Feb", "Mar", "Apr"]
#x_axis = [1, 2, 3]
#y_axis = [5, 10, 15]
#x_axis = weekly_totals_df["date"]
plt.figure(figsize=(20, 6))
# Create the plots.
plt.plot(weekly_totals_df["Rural"], color="cornflowerblue", linewidth=4, label='Rural')
plt.plot(weekly_totals_df["Suburban"], color="red", linewidth=4, label='Suburban')
plt.plot(weekly_totals_df["Urban"], color="gold", linewidth=4, label='Urban')


# Create labels for the y axes.
plt.ylabel("Fare ($USD)")
# Set the y limit between 0 and 45.
plt.ylim(-50, 2600)

# change the dates to months so it matches the example
plt.xticks(["2019-01", "2019-02", "2019-03", "2019-04", "2019-05"], ['Jan', 'Feb', 'Mar', "Apr", ""])
#plt.xticks(np.arange(4), ["Jan", "Feb", "Mar", "Apr"])
# Create a title.
plt.title("Total Fare by City Type", fontsize = 20)
# Add a grid.
plt.grid(True)

#plt.margins(0.2)
# Add the legend.
plt.legend(title="type", loc="center")
plt.savefig("analysis/Pyber_fare_summary.png")
```
